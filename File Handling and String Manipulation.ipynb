 # Task 2

Modifying data

```python
conv_f = open("conv.txt", encoding = "utf-8")
lines = conv_f.readlines()

s_conv = []

for line in lines:
    if(line=="\n"):
        continue
    s_conv.append(line[:-2])
```

```python
s_conv
```

```python
for i in range(len(s_conv)):
    s_conv[i] = s_conv[i].lower()
s_conv
```

Cleaning data

```python
char1 = "’"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char1, '')
char2 = ","
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char2, '')

char3 = "?"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char3, '')

char4 = "…"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char4,'')

char5 = "ive"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char5, 'i have ')
    
char6 = "wont"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char6, 'will not ')
    
char7 = "fathers"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char7, 'father')

char8 = "mothers"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char8, 'mother')

char9 = "hes"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char9,'he is')

char10 = "im"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char10, 'i am')
    
char11 = "its"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char11, 'it is')
    
char12 = "dont"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char12, 'do not')
    
char13 = "nights"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char13, 'night')

char14 = "thats"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char14, 'that is')

char15 = "youve"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char15,'you have')

char16 = "weve"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char16, 'we have')
    
char17 = "havent"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char17, 'have not')
    
char18 = "youre"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char18, 'you are')
    
char19 = "were"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char19, 'we are')
    
char20 = "wheres"
for p,q in enumerate(s_conv):
    s_conv[p] = q.replace(char20, 'where is')
    
s_conv
```

```python
a = 0
Dialogues = []
while a != 111:
    b = s_conv[:][a]
    c = b.split(":")
    Dialogues.append(c)
    a = a+1
    continue
```

```python
len(Dialogues)
```

```python
unique_name = dict()

for i,j in Dialogues:
    if i in unique_name.keys():
        unique_name[i].append(j)
    else:
        unique_name[i]=[j]
```

```python
unique_name.keys()
```

```python
keys = []

for i in unique_name.keys():
    keys.append(i)
    
dlg = []

for speakers in keys:
    s = unique_name[speakers]
    dlg.append(s)
```
